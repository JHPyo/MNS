```python
#TTL 파일 (_0.csv)을 CS on 동안만을 편집하는 코드 -> CS on 일 때 세번째 colounm이 1임
import pandas as pd
import os

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError: 
        print ('Error: Creating directory.' + directory)

#path =  filedialog.askdirectory(initialdir = "/", title = "Select directory")

#아래에 r'원하는 경로'로 입력 
path =r'F:\Data\Two-way active avoidance behavior\Experiments 별 정리\pBLA Ppp1r1b GRAB DA\L134\vid'
print (path)  # 원하는 xlsx파일들이 포함된 폴더를 선택 #

createFolder(path[:-4]+'/TTL/')

for root, dirs, files in os.walk(path):
    for fname in files:
        if('.csv' in fname): 

            full_name = os.path.join(root, fname)
            print('In Processing :' + full_name)
            data = pd.read_csv(full_name, header = 1)
            
            #TTL이 없는 행 삭제 
            data2 = data[data.iloc[:,5] != 0]
        
            Start = []
            End = []
            Event = []
            
            
            Start.append(data2.iloc[0,0])
            for i in range(len(data2.iloc[:,0])):
                if(data2.iloc[i,0] - data2.iloc[i-1,0] > 0.5):
                    Start.append(data2.iloc[i,0])
                    End.append(data2.iloc[i-1,0])
            End.append(data2.iloc[-1,0])
            for i in range(len(Start)):
                if End[i]-Start[i] > 24.99:
                    Event.append('Escape+')
                elif End[i]-Start[i] < 24.99:
                    Event.append('Avoidance+')
    
                df = pd.DataFrame(data = list(zip(Event, Start, End)), columns = ['Event','onset', 'offset'])
        
                df.to_csv(path[:-4]+'/TTL/'+fname[:-8]+'TTL.csv', index = 0)
```
